```python

data2 = data.copy()
```

```python
#### replace null values with mean
```

```python
data2 = data2.fillna(data2.mean())
```

```python
data2
```

```python
data2.info()
```

```python
data2.columns
```

```python
dist=(data2['AQI'])
distset=set(dist)
dd=list(distset)
dictofwords={dd[i] : i for i in range(0, len(dd))}
data2['AQI']=data2['AQI'].map(dictofwords)
```

```python
data2
```

```python
data2.tail()
```

```python
data2.isnull().sum()
```

```python
IAQ_bins = [0,50,100,200,300,400,500]
IAQ_labels = ["Good", "Satisfactory", "Moderate", "poor","Very poor","Severe"]
data2['IAQc'] = pd.cut(data2['IAQ'], IAQ_bins, labels=IAQ_labels, 
 right=True, include_lowest=True)
data2
```

```python
data2=data2.drop('IAQ',1)
data2=data2.drop('AQI',1)

```

```python
data2.columns
```

```python
features=data2.drop(['IAQc'], axis=1)
labels=data2['IAQc']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(features,labels,test_size=0.2,random_state=2)
```
